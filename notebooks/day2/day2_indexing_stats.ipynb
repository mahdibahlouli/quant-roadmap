```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
tickers = ["BTC","ETH","SOL"]
data = {}
for t in tickers :
    df = pd.read_csv(f"../../res/{t}_daily_historical_data.csv", parse_dates=["Date"], index_col="Date")
    data[t] = df
```

```python
print(data["SOL"])
```

```python

# show the rows 
print(data["SOL"].loc["2025-09"])
# show the Series
print(data["SOL"]["Close/Last"])
# show a range of dates 
print(data["SOL"].iloc[0:5])
```

```python
# show the rows 
print(data["SOL"].loc["2025-08"])
# show a range of dates 
print(data["SOL"].iloc[0:12])
```

```python

# show the Series
print(data["SOL"]["Close/Last"])
```

```python
# now to create a dataframe where we can put the returns 
returns = pd.DataFrame()

for t in tickers:
    returns[t] = data[t]["Close/Last"].pct_change()

print(returns)
    
```

```python
# my first intuition of showing the stats 
stats = returns.describe()
print(stats)
```

```python
# the dynamic methode for showing the stas 
stats = returns.agg(["mean","std","min","max"])
print(stats)
```

```python
# my first intuition of plotting the close price of all the tickers 
for t in tickers:
    data[t]["Close/Last"].plot()
    plt.show()
```

```python
plt.figure(figsize=(12,6))
for t in tickers: 
    plt.plot(data[t]["Close/Last"], label = t)
plt.title("closing prices of btc,eth,sol")
plt.legend()
plt.show()
```
