```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df= pd.read_csv("../../res/BTC_daily_historical_data.csv", parse_dates=["Date"], index_col="Date")
df.head(10)
```

```python
df["Daily Returns"] = df["Close/Last"].pct_change()
df.head(10)
```

```python
df["Daily Returns"].plot(x="Date",y="Daily Returns")
plt.show()
```
